```python
from bs4 import BeautifulSoup as soup
import requests

def scrape_page(url):
    print("Scraping page:", url)
    html = requests.get(url)
    bsobj = soup(html.content, 'html.parser')
    
    articles = bsobj.find_all('div', {'class':"img-context"})
    
    headlines = {}
    for article in articles:
        headline_tag = article.find("h2")
        if headline_tag:
            headline_text = headline_tag.text.strip()
        else:
            headline_text = ""
        
        content_tag = article.find("p")
        if content_tag:
            content_text = content_tag.text.strip()
        else:
            content_text = ""
        
        date_tag = article.find("div", class_="date")
        if date_tag:
            date_text = date_tag.text.strip()
        else:
            date_text = ""
        
        headlines[headline_text] = {
            "content": content_text,
            "date": date_text
        }
    
    return headlines

base_url = 'https://indianexpress.com/section/political-pulse/'

all_headlines = {}

for page_num in range(1, 700):  
    url = base_url + f'page/{page_num}/'
    
    headlines = scrape_page(url)
    
    all_headlines.update(headlines)
    time.sleep(5)

for headline, data in all_headlines.items():
    print("Headline:", headline)
    print("Content:", data["content"])
    print("Date:", data["date"])
    print("-" * 50)
```
